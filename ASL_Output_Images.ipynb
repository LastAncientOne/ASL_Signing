# American Sign Language (ASL) Loading Images

```python
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
from tqdm import tqdm
```

```python
asl_data = "C:/ASL_Project/ABCImages"
```

```python
ABC = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'K', 'L', 'M', 'N', 'O', 'P' ,'Q', 'R', 'S', 'T', 'U', 'V', 'W', 'X', 'Y']
```

```python
for abc in ABC:  # Loop ABC
    path = os.path.join(asl_data, abc)  # create path 
    for img in os.listdir(path):  # iterate over each image per ABC
        img_array = cv2.imread(os.path.join(path,img))  # convert to array
        plt.imshow(img_array)  # graph it
        plt.show()  # display images
```
